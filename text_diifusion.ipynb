```python
import sys
import logging
from packaging import version

logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s",
    handlers=[logging.StreamHandler(sys.stdout)]
)
logger = logging.getLogger(__name__)

def validate_environment():
    """Verify all dependencies and environment configurations"""
    try:
        # Check Python version
        if sys.version_info < (3, 8):
            raise RuntimeError("Python 3.8 or higher required")

        # Verify NumPy availability and version
        try:
            import numpy as np
            np_version = np.__version__
            if version.parse(np_version) < version.parse("1.21"):
                logger.warning(f"NumPy {np_version} detected. Consider upgrading")
        except ImportError:
            raise ImportError("NumPy not found. Please install with: !pip install numpy")

        # Verify PyTorch installation
        try:
            import torch
            if not torch.cuda.is_available():
                logger.warning("CUDA not available - using CPU (may be slow)")
        except ImportError:
            raise ImportError("PyTorch not found. Install with: !pip install torch")

        # Verify Diffusers installation
        try:
            import diffusers
            if version.parse(diffusers.__version__) < version.parse("0.19.0"):
                logger.warning("Diffusers version may be outdated")
        except ImportError:
            raise ImportError("Diffusers not found. Install with: !pip install diffusers")

    except Exception as e:
        logger.error(f"Environment validation failed: {str(e)}")
        raise

# Perform environment validation at startup
validate_environment()

import torch
import numpy as np
from diffusers import StableDiffusionPipeline
from diffusers.schedulers import PNDMScheduler
from PIL import Image

def create_pipeline():
    """Create and configure the diffusion pipeline with optimizations"""
    try:
        device = "cuda" if torch.cuda.is_available() else "cpu"
        dtype = torch.float16 if device == "cuda" else torch.float32

        # Use optimized scheduler configuration
        scheduler = PNDMScheduler(
            beta_start=0.00085,
            beta_end=0.012,
            beta_schedule="scaled_linear",
            num_train_timesteps=1000
        )

        pipe = StableDiffusionPipeline.from_pretrained(
            "runwayml/stable-diffusion-v1-5",
            scheduler=scheduler,
            torch_dtype=dtype,
            use_safetensors=True
        ).to(device)

        # Enable memory optimizations
        if device == "cuda":
            pipe.enable_attention_slicing()
            pipe.enable_vae_slicing()

        return pipe, device

    except Exception as e:
        logger.error(f"Pipeline creation failed: {str(e)}", exc_info=True)
        raise

def generate_image(pipe, device, prompt, guidance_scale=7.5, steps=50):
    """Generate image with comprehensive error handling"""
    try:
        with torch.autocast(device_type=device, enabled=True):
            image = pipe(
                prompt=prompt,
                guidance_scale=guidance_scale,
                num_inference_steps=steps
            ).images[0]

        return image

    except RuntimeError as e:
        if "CUDA out of memory" in str(e):
            logger.error("CUDA memory error - try reducing image dimensions")
        elif "Numpy" in str(e):
            logger.error("NumPy operation failed - verify installation")
        else:
            logger.error(f"Generation failed: {str(e)}", exc_info=True)
        raise

def main():
    """Main execution with full error handling"""
    try:
        # Validate environment again before execution
        validate_environment()

        # Create pipeline
        pipe, device = create_pipeline()

        # Generation parameters
        prompt = "A cyberpunk cityscape with neon lights and flying cars, ArtStation quality"
        guidance_scale = 7.5
        steps = 50

        # Generate image
        image = generate_image(pipe, device, prompt, guidance_scale, steps)

        # Save and display
        timestamp = np.datetime64('now').astype(str).replace(":", "-")
        filename = f"generated_image_{timestamp}.png"
        image.save(filename)
        logger.info(f"Image saved as {filename}")

        # Display in notebooks
        try:
            from IPython.display import display
            display(image)
        except ImportError:
            logger.info("IPython not available - skipping display")

    except KeyboardInterrupt:
        logger.warning("Process interrupted by user")
    except Exception as e:
        logger.error(f"Fatal error: {str(e)}", exc_info=True)
        sys.exit(1)

if __name__ == "__main__":
    main()
```

```python
from diffusers import StableDiffusionPipeline, StableDiffusionImg2ImgPipeline
import torch
from IPython.display import display
from PIL import Image
import os
from google.colab import drive
import logging
from datetime import datetime
import sys

# Configure logging for detailed progress tracking
logging.basicConfig(level=logging.INFO, format="%(asctime)s - %(levelname)s - %(message)s")
logger = logging.getLogger(__name__)

# Mount Google Drive and set save path
try:
    drive.mount('/content/drive')
    save_path = "/content/drive/MyDrive/Colab_Projects/stable_diffusion"
    os.makedirs(save_path, exist_ok=True)
    logger.info(f"Save directory set to: {save_path}")
except Exception as e:
    logger.error(f"Failed to mount Google Drive or create directory: {str(e)}")
    raise

# Define prompts for generation
online_prompts = [
    "A cyberpunk cityscape with neon lights and flying cars, trending on ArtStation",
    "A majestic dragon soaring through a fantasy landscape",
]

offline_prompts = [
    (
        "/content/drive/MyDrive/Colab_Projects/stable_diffusion/im1.jpg",
        "A futuristic version of the input image with neon lighting"
    ),
    (
        "/content/drive/MyDrive/Colab_Projects/stable_diffusion/imges1.jpg",
        "A surreal reinterpretation of the input image with glowing elements"
    )
]

# Select device with fallback to CPU
device = "cuda" if torch.cuda.is_available() else "cpu"
logger.info(f"Using device: {device}")
if device == "cpu":
    logger.warning("Running on CPU may result in slower performance.")

# Load pipelines with optimization and specific error handling
try:
    pipe = StableDiffusionPipeline.from_pretrained(
        "runwayml/stable-diffusion-v1-5",
        torch_dtype=torch.float16,  # Reduced memory usage
        use_safetensors=True        # Faster loading, lower memory footprint
    ).to(device)

    img2img_pipe = StableDiffusionImg2ImgPipeline.from_pretrained(
        "runwayml/stable-diffusion-v1-5",
        torch_dtype=torch.float16,
        use_safetensors=True
    ).to(device)
    logger.info("Pipelines loaded successfully.")
except OSError as e:
    logger.error(f"Failed to load model files: {str(e)}")
    raise
except Exception as e:
    logger.error(f"Unexpected error loading pipelines: {str(e)}")
    raise

def generate_and_save_image(
    text_prompt=None,
    image_path=None,
    filename="output",
    strength=0.75,
    guidance_scale=7.5,
    num_inference_steps=50,
    seed=None
):
    """
    Generate and save an image using Stable Diffusion pipelines.

    Args:
        text_prompt (str): Text prompt for generation.
        image_path (str, optional): Path to input image for image-to-image generation.
        filename (str): Base name for the output file.
        strength (float): Transformation strength for img2img (0.0 to 1.0).
        guidance_scale (float): Influence of the text prompt (non-negative).
        num_inference_steps (int): Number of denoising steps (positive integer).
        seed (int, optional): Random seed for reproducibility.

    Raises:
        ValueError: If parameters are invalid or inputs are missing/inappropriate.
        FileNotFoundError: If image_path is invalid or file is not found.
        IOError: If image cannot be loaded or saved.
        Exception: For unexpected errors during generation.
    """
    try:
        # Parameter validation
        if strength is not None and not (0 <= strength <= 1):
            raise ValueError("Strength must be between 0 and 1")
        if guidance_scale < 0:
            raise ValueError("Guidance scale must be non-negative")
        if num_inference_steps <= 0 or not isinstance(num_inference_steps, int):
            raise ValueError("Number of inference steps must be a positive integer")

        # Input validation
        if image_path:
            if not os.path.isfile(image_path):
                raise FileNotFoundError(f"Image file not found at {image_path}. Verify path and upload status.")
            if not text_prompt:
                raise ValueError("Text prompt is required for image-to-image generation")
            logger.info(f"Processing image-to-image generation for {image_path}")
        else:
            if not text_prompt:
                raise ValueError("Text prompt is required for text-to-image generation")
            logger.info(f"Processing text-to-image generation with prompt: {text_prompt}")

        # Set seed for reproducibility
        if seed is None:
            seed = torch.randint(0, 2**32 - 1, (1,)).item()
            logger.info(f"No seed provided, using random seed: {seed}")
        else:
            logger.info(f"Using seed: {seed}")
        torch.manual_seed(seed)

        # Log generation parameters
        logger.info(f"Parameters: prompt='{text_prompt}', image_path='{image_path}', "
                    f"strength={strength}, guidance_scale={guidance_scale}, "
                    f"num_inference_steps={num_inference_steps}, seed={seed}")

        # Use autocast and inference_mode for efficiency
        with torch.autocast(device), torch.inference_mode():
            if image_path:
                # Load and preprocess input image
                try:
                    init_image = Image.open(image_path).convert("RGB")
                    init_image = init_image.resize((512, 512), Image.Resampling.LANCZOS)
                except (IOError, Image.UnidentifiedImageError) as e:
                    logger.error(f"Failed to load image: {str(e)}")
                    raise IOError(f"Invalid image file: {str(e)}")
                image = img2img_pipe(
                    prompt=text_prompt,
                    image=init_image,
                    strength=strength,
                    guidance_scale=guidance_scale,
                    num_inference_steps=num_inference_steps
                ).images[0]
            else:
                image = pipe(
                    prompt=text_prompt,
                    guidance_scale=guidance_scale,
                    num_inference_steps=num_inference_steps,
                    height=512,
                    width=512
                ).images[0]

        # Create unique filename with timestamp
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        base_name, ext = os.path.splitext(filename)
        if not ext:
            ext = ".png"
        unique_filename = f"{base_name}_{timestamp}{ext}"
        full_path = os.path.join(save_path, unique_filename)

        # Save image with optimized compression
        try:
            image.save(full_path, format="PNG", compress_level=1)
            logger.info(f"Success: Image saved to {full_path}")
        except Exception as e:
            logger.error(f"Failed to save image: {str(e)}")
            raise IOError(f"Could not save image: {str(e)}")

        # Display image only in IPython environments
        if 'IPython' in sys.modules:
            display(image)
        else:
            logger.info("Image generated but not displayed (not in IPython environment)")

        # Clean up GPU memory
        if device == "cuda":
            torch.cuda.empty_cache()
            logger.info("CUDA memory cache cleared.")

    except (ValueError, FileNotFoundError, IOError) as e:
        logger.error(f"Error: {str(e)}")
        raise
    except Exception as e:
        logger.error(f"Unexpected error during generation: {str(e)}")
        raise

# Generate images from online prompts
for idx, prompt in enumerate(online_prompts):
    generate_and_save_image(
        text_prompt=prompt,
        filename=f"online_{idx}",
        guidance_scale=5.0,
        num_inference_steps=50,
        seed=42  # Ensure reproducibility
    )

# Generate images from offline prompts
for idx, (img_path, txt_prompt) in enumerate(offline_prompts):
    generate_and_save_image(
        text_prompt=txt_prompt,
        image_path=img_path,
        filename=f"offline_{idx}",
        strength=0.75,
        guidance_scale=5.0,
        num_inference_steps=50,
        seed=42  # Ensure reproducibility
    )
```

```python

from diffusers import StableDiffusionPipeline, StableDiffusionImg2ImgPipeline
import torch
import gradio as gr
from PIL import Image
import os
from google.colab import drive
import logging
from datetime import datetime

# Configure logging for detailed progress tracking
logging.basicConfig(level=logging.INFO, format="%(asctime)s - %(levelname)s - %(message)s")
logger = logging.getLogger(__name__)

# Mount Google Drive and set save path
try:
    drive.mount('/content/drive', force_remount=True)
    save_path = "/content/drive/MyDrive/Colab_Projects/stable_diffusion"
    os.makedirs(save_path, exist_ok=True)
    logger.info(f"Save directory set to: {save_path}")
except Exception as e:
    logger.error(f"Failed to mount Google Drive: {str(e)}")
    raise RuntimeError("Failed to mount Google Drive. Check authorization.") from e

# Verify image paths
image_paths = [
    "/content/drive/MyDrive/Colab_Projects/stable_diffusion/im1.jpg",
    "/content/drive/MyDrive/Colab_Projects/stable_diffusion/imges1.jpg"
]
for img_path in image_paths:
    if not os.path.exists(img_path):
        logger.error(f"File not found: {img_path}")
        raise FileNotFoundError(f"Upload missing file to: {img_path}")

# Select device with fallback to CPU
device = "cuda" if torch.cuda.is_available() else "cpu"
logger.info(f"Using device: {device}")
if device == "cpu":
    logger.warning("Running on CPU may result in slower performance.")

# Global pipeline variables
pipe = None
img2img_pipe = None

def load_models():
    """Load pipelines only once to optimize memory and time."""
    global pipe, img2img_pipe
    if not pipe:
        try:
            pipe = StableDiffusionPipeline.from_pretrained(
                "runwayml/stable-diffusion-v1-5",
                torch_dtype=torch.float16,
                use_safetensors=True  # Faster loading and lower memory footprint
            ).to(device)
            logger.info("Text-to-Image pipeline loaded.")
        except Exception as e:
            logger.error(f"Failed to load Text-to-Image pipeline: {str(e)}")
            raise
    if not img2img_pipe:
        try:
            img2img_pipe = StableDiffusionImg2ImgPipeline.from_pretrained(
                "runwayml/stable-diffusion-v1-5",
                torch_dtype=torch.float16,
                use_safetensors=True
            ).to(device)
            logger.info("Image-to-Image pipeline loaded.")
        except Exception as e:
            logger.error(f"Failed to load Image-to-Image pipeline: {str(e)}")
            raise

def generate_text_to_image(prompt, guidance_scale, steps):
    """
    Generate an image from a text prompt.

    Args:
        prompt (str): Text prompt for generation.
        guidance_scale (float): Influence of the text prompt.
        steps (int): Number of denoising steps.

    Returns:
        PIL.Image: Generated image or error message.
    """
    try:
        if not prompt:
            raise ValueError("Prompt cannot be empty.")
        load_models()
        with torch.autocast(device), torch.inference_mode():
            image = pipe(
                prompt,
                guidance_scale=guidance_scale,
                num_inference_steps=steps
            ).images[0]
        filename = f"text2img_{prompt[:10]}_{steps}steps_{datetime.now().strftime('%Y%m%d_%H%M%S')}.png"
        full_path = os.path.join(save_path, filename)
        image.save(full_path, format="PNG", compress_level=1)
        logger.info(f"Text-to-Image generated and saved to {full_path}")
        return image
    except ValueError as e:
        logger.error(f"Input error: {str(e)}")
        return f"Error: {str(e)}"
    except Exception as e:
        logger.error(f"Unexpected error in Text-to-Image: {str(e)}")
        return f"Error: {str(e)}"

def generate_image_to_image(input_image, prompt, strength, guidance_scale, steps):
    """
    Generate an image from an input image and text prompt.

    Args:
        input_image (PIL.Image): Input image for transformation.
        prompt (str): Text prompt for guiding transformation.
        strength (float): Transformation strength (0.0 to 1.0).
        guidance_scale (float): Influence of the text prompt.
        steps (int): Number of denoising steps.

    Returns:
        PIL.Image: Generated image or error message.
    """
    try:
        if not input_image:
            raise ValueError("Input image is required.")
        if not prompt:
            raise ValueError("Prompt cannot be empty.")
        if not (0 <= strength <= 1):
            raise ValueError("Strength must be between 0 and 1.")
        load_models()
        input_image = input_image.resize((512, 512)).convert("RGB")
        with torch.autocast(device), torch.inference_mode():
            image = img2img_pipe(
                prompt=prompt,
                image=input_image,
                strength=strength,
                guidance_scale=guidance_scale,
                num_inference_steps=steps
            ).images[0]
        filename = f"img2img_{prompt[:10]}_{steps}steps_{datetime.now().strftime('%Y%m%d_%H%M%S')}.png"
        full_path = os.path.join(save_path, filename)
        image.save(full_path, format="PNG", compress_level=1)
        logger.info(f"Image-to-Image generated and saved to {full_path}")
        return image
    except ValueError as e:
        logger.error(f"Input error: {str(e)}")
        return f"Error: {str(e)}"
    except Exception as e:
        logger.error(f"Unexpected error in Image-to-Image: {str(e)}")
        return f"Error: {str(e)}"

# Gradio Interface
with gr.Blocks() as demo:
    gr.Markdown("# Stable Diffusion Playground")

    with gr.Tab(" Text-to-Image"):
        with gr.Row():
            text_prompt = gr.Textbox(label="Describe your vision", placeholder="e.g., 'A cyberpunk city with neon lights'")
            text_output = gr.Image(label="Generated Artwork")
        with gr.Row():
            guidance_slider = gr.Slider(5.0, 15.0, 7.5, label="Guidance Scale (Creativity vs. Accuracy)")
            steps_slider = gr.Slider(25, 100, 50, label="Quality Steps (Higher = Slower)")
        text_generate_btn = gr.Button("Generate Masterpiece")

    with gr.Tab(" Image-to-Image"):
        with gr.Row():
            input_image = gr.Image(label="Upload Base Image", type="pil")
            img_output = gr.Image(label="Modified Artwork")
        with gr.Row():
            img_prompt = gr.Textbox(label="Describe modifications", placeholder="e.g., 'Add glowing elements'")
            strength_slider = gr.Slider(0.0, 1.0, 0.75, label="Transformation Strength")
        with gr.Row():
            img_guidance_slider = gr.Slider(5.0, 15.0, 7.5, label="Guidance Scale")
            img_steps_slider = gr.Slider(25, 100, 50, label="Quality Steps")
        img_generate_btn = gr.Button("Transform Image")

    # Connect buttons to functions
    text_generate_btn.click(
        generate_text_to_image,
        inputs=[text_prompt, guidance_slider, steps_slider],
        outputs=text_output
    )
    img_generate_btn.click(
        generate_image_to_image,
        inputs=[input_image, img_prompt, strength_slider, img_guidance_slider, img_steps_slider],
        outputs=img_output
    )

if __name__ == "__main__":
    try:
        demo.queue().launch(debug=True, share=True)
    except Exception as e:
        logger.error(f"Failed to launch Gradio interface: {str(e)}")
        raise
```
